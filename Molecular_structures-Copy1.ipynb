# molecular Structures of chemicals present in GRPF 

#### Methyl Ethyl Ketone Peroxide

```python
from rdkit import Chem 
m = Chem.MolFromSmiles('O([C@@H](C(COC(=O)C(C)C)(C)C)C(C)C)C(=O)C(C)C') 
m
```

#### Dimethyl Phthalate

```python
m = Chem.MolFromSmiles('O(C(=O)c1c(cccc1)C(=O)OC)C') 
m
```

#### 2,2,4-Trimethyl-1,3-pentanediol diisobutyrate

```python
m = Chem.MolFromSmiles('O([C@@H](C(COC(=O)C(C)C)(C)C)C(C)C)C(=O)C(C)C')
m
```

#### Hydrogen Peroxide

```python
m = Chem.MolFromSmiles('OO')
m
```

#### Methyl Ethyl Ketone


```python
m = Chem.MolFromSmiles('O=C(CC)C')
m
```

#### Water

```python
m = Chem.MolFromSmiles('O')
m
```

#### Antimony Trioxide


```python
m = Chem.MolFromSmiles('[Sb+3].[Sb+3].[O-2].[O-2].[O-2]')
m
```

#### Cobalt Octoate

```python
m = Chem.MolFromSmiles('OC(=O)[C@@H](CCCC)CC')
m
```

#### Styrene


```python
m = Chem.MolFromSmiles('c1(ccccc1)C=C')
m
```

#### Maleic Anhydride


```python
m = Chem.MolFromSmiles('O1C(=O)C=CC1=O')
m
```

#### Fumed Silicia (CAB-O-SIL)


```python
m = Chem.MolFromSmiles('[Si](=O)=O')
m
```
